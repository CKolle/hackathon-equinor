```python
import requests
import pandas as pd
from datetime import datetime, timedelta

# Define the start and end dates
start_date = datetime(2013, 1, 1)
end_date = datetime(2014, 12, 31)

# Initialize an empty list to store data
all_data = []

# Loop through each date
current_date = start_date
while current_date <= end_date:
    date_str = current_date.strftime("%Y-%m-%d")  # Format date as YYYY-MM-DD
    print(date_str)
    url = f"https://www.yr.no/api/v0/locations/1-72837/observations/{date_str}"

    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise an error for failed requests
        data = response.json()
        
        # Extract hourly observations if available
        if "historical" in data and "days" in data["historical"] and data["historical"]["days"]:
            for hour in data["historical"]["days"][0]["hours"]:
                all_data.append({
                    "date": date_str,
                    "time": hour["time"],
                    "temperature": hour["temperature"]["value"] if "temperature" in hour and "value" in hour["temperature"] else None,
                    "wind_speed": hour["wind"]["meanSpeed"] if "wind" in hour and "meanSpeed" in hour["wind"] else None
                })
    except:
        print(f"Failed to fetch data for {date_str}")

    # Move to the next day
    current_date += timedelta(days=1)

# Convert the collected data into a DataFrame
df = pd.DataFrame(all_data)

# Save to CSV file
df.to_csv("weather_data_2010_2014.csv", index=False)

print("Download complete. Data saved to 'weather_data_2010_2014.csv'")
```
